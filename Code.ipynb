```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bank_data = pd.read_csv("bank-full.csv",sep = ';')
bank_data.head()
```

Q1. Factors influencing the purchase of a bank product.

```python
# To Chech how many have taken the deposit
value_counts = bank_data['y'].value_counts()
value_counts
vc =value_counts.plot(kind='bar')
for p in vc.patches:
    vc.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
```

```python
#job vs deposit
job_df = pd.DataFrame()

job_df['yes'] = bank_data[bank_data['y'] == 'yes']['job'].value_counts()
job_df['no'] = bank_data[bank_data['y'] == 'no']['job'].value_counts()

jb=job_df.plot(kind='bar')
for p in jb.patches:
    jb.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    plt.xlabel('Job', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.title("Job vs Deposit")
```

```python
#education and deposit
education_df = pd.DataFrame()

education_df['yes'] = bank_data[bank_data['y'] == 'yes']['education'].value_counts()
education_df['no'] = bank_data[bank_data['y'] == 'no']['education'].value_counts()
ax=education_df.plot(kind='bar')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    plt.xlabel('Education ', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.title("Education  vs Deposit")
```

```python
#Marital and deposit
Marital_df = pd.DataFrame()

Marital_df['yes'] = bank_data[bank_data['y'] == 'yes']['marital'].value_counts()
Marital_df['no'] = bank_data[bank_data['y'] == 'no']['marital'].value_counts()
ax=Marital_df.plot(kind='bar')
for p in ax.patches:
    ax.annotate(str(p.get_height()), (p.get_x() * 1.005, p.get_height() * 1.005))
    plt.xlabel('Marital', fontsize=14)
    plt.ylabel('Count', fontsize=14)
    plt.title("Marital vs Deposit")
```

Q3. To find the average duration time of customers who have subscribed and not subscribed the bank term.

```python
print("Median of duration who has subscribed: ", bank_data[bank_data['y']=='yes']['duration'].median())
print("Median of duration who hasn't subscribed: ", bank_data[bank_data['y']=='no']['duration'].median())
print("Summary of duration who hasn't subscribed: ", bank_data[bank_data['y']=='no']['duration'].describe())
print("Summary of duration who has subscribed: ", bank_data[bank_data['y']=='yes']['duration'].describe())
```
